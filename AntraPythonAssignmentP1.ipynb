##### Define functions to perform all the necessary cleaning step

##### Create a function to format all the names

```python
def NameCleaning(dflist):
    list = []
    for element in dflist:
        element = element.strip()
        FirstLetter = element[0]
        OtherLetter = element[1:]
        FirstLetter = FirstLetter.upper()
        OtherLetter = OtherLetter.lower()
        element = FirstLetter + OtherLetter
        list.append(element)
    return list
```

##### Create a function to format all the phone numbers

```python
def PhoneCleaning(dflist):
    list = []
    for number in dflist:
        number = number.strip()
        number = number.replace("-", "")
        # assert (len(number) == 9), "not valid phone number!"
        # this line generating trouble tried to trouble shoot
        #if len(number) != 9:
        #    print(str(number) + " " + str(dflist[number]))
        number = number[-8:]
        list.append(number)
    return list
```

##### Create a function to format Address

```python
def AddressCleaning(dflist):
    list = []
    for address in dflist:
        address = address.strip()
        address = address.replace("No.", "")
        address = address.replace("No", "")
        address = address.replace("#", "")
        list.append(address)
    return list
```

##### Perform the transformation

```python
df1["FirstName"] = NameCleaning(df1["FirstName"])
df2["FirstName"] = NameCleaning(df2["FirstName"])
df1["LastName"] = NameCleaning(df1["LastName"])
df2["LastName"] = NameCleaning(df2["LastName"])
```

```python
df1["Address"] = AddressCleaning(df1["Address"])
df2["Address"] = AddressCleaning(df2["Address"])
```

```python
df1["Phone"] = PhoneCleaning(df1["Phone"])
df2["Phone"] = PhoneCleaning(df2["Phone"])
```
